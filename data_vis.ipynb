```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("results_test.csv")
```

```python
df["train_acc"].str.split("(").str.get(1).str[:-1]
```

```python
df["train_acc"]=df["train_acc"].str.split("(").str.get(1).str[:-1].astype("float64")
df["val_acc"]=df["val_acc"].str.split("(").str.get(1).str[:-1].astype("float64")
```

```python
df
```

```python
plt.plot(df['epoch'],df['train_loss'],label='train')
plt.plot(df['epoch'],df['val_loss'],label='validation')
plt.legend()
plt.title('Loss')
plt.show()


plt.plot(df['epoch'],df['train_acc'],label='train')
plt.plot(df['epoch'],df['val_acc'],label='validation')
plt.legend()
plt.title("Accuracy")
plt.show()
```
